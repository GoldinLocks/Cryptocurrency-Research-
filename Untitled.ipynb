```python
import warnings
warnings.filterwarnings('ignore')

import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import matplotlib as mpl
import matplotlib.cm as cm
import matplotlib.patches as mpatches
import seaborn as sns

import ffn

from math import sqrt
from scipy import stats

import cvxpy
from tqdm import tqdm, trange

mpl.style.use('seaborn')
%matplotlib inline
```

```python
rets9 = pd.read_excel('strat1-rets.xlsx')
rets9 = pd.DataFrame(rets9).dropna()

rets9.index = rets9['Date/Time']
rets9.index = pd.to_datetime(rets9.index)
#rets9 = rets9.drop(['Date/Time','Price', 'Contracts'], axis=1)
rets9 = rets9.drop(['Date/Time'], axis=1)
rets9 = rets9.dropna(axis=1, how='any', thresh=None, subset=None, inplace=False)
rets9 = rets9.astype(float)
rets9['Price_ch'] = rets9['Price'].pct_change()
rets9['Returns'] = rets9['Price_ch'] + rets9['Contracts']
rets9 = rets9.dropna()
rets9.head()
```

```python
def rolling_sharpe(df, window):
    return df.rolling(window=window, center=False).mean()/df.rolling(window=window, center=False).std() * sqrt(252)
```

```python
subsample_rate = 21
trad_rolling_sharpe = rolling_sharpe(rets9.Returns, subsample_rate)
```

```python
rets981 = pd.read_excel('newstat.xlsx')
rets981 = pd.DataFrame(rets981).fillna(0)
rets981.columns = ['Perf Metric', 'All', 'Long', 'Short']
rets981.head()
```

```python
plt.subplot(411)
rets9.Profit.cumsum().plot(figsize=(9,14), color='#28c940')
plt.title('Portfolio Returns')
plt.ylabel('USD($)')
plt.legend(loc='lower right');

plt.subplot(412)
ffn.to_drawdown_series(rets9.Profit).plot(kind = 'area',alpha=.4,label='Drawdown',color='#ff3838',figsize=(9,14),grid=False)
plt.title('Drawdowns')
plt.ylabel('Drawdown (USD$)')
plt.legend(loc='lower right')

plt.subplot(413)
trad_rolling_sharpe.plot(label='sharpe',figsize=(9,14))
plt.axhline(y=trad_rolling_sharpe.mean(), color='red', linewidth=1.5, linestyle='dashed', label='mean')
plt.margins(0.02)
plt.legend()
plt.title('Portfolio 21D Rolling Sharpe Ratio')
plt.ylabel('Sharpe Ratio')
plt.legend(loc='upper right')

plt.tight_layout()
plt.show();

sharpe = ((rets9.Profit.mean() / rets9.Profit.std()) * sqrt(252))
print('             Portfolio 21D Rolling Sharpe =',sharpe)

daily_rets9 = rets9.resample('BMS').last()
daily_rets9 = daily_rets9.drop(['Contracts','Price', 'Price_ch', 'Returns'], axis=1)
daily_rets9 = daily_rets9.dropna(axis=1, how='any', thresh=None, subset=None, inplace=False)

plt.figure(figsize=(14, 3))
sns.heatmap(daily_rets9.T, cmap="RdYlGn", annot=True)
plt.title('Monthly Returns')
plt.show();

rets981
```
